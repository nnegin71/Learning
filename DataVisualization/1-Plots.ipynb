```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 
```

```python
# loading data
fifa_data = pd.read_csv('./Data/fifa.csv', index_col = 'Date', parse_dates=True)
fifa_data.head()
```

```python
fifa_data.tail()
```

# Line Plot

```python
# plot the data
plt.figure(figsize=(14,6))
plt.title('Fifa data')
sns.lineplot(fifa_data)
plt.xlabel('Date')
```

```python
# plot a subset of data
fifa_data.columns
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(fifa_data.GER,label ='Germany')
sns.lineplot(fifa_data.FRA,label ='France')
plt.legend()
plt.title('Fifa data')
plt.xlabel('Date')
```

# Bar Plot

```python
flight_data = pd.read_csv('./Data/flight_delays.csv', index_col='Month')
flight_data
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=flight_data.index , y=flight_data['AA'])
plt.ylabel('Average Flight Delay (min)')
plt.xlabel('Month')
```

# HeatMap

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_data, annot=True)
plt.ylabel('Month')
plt.xlabel('Airlines')
```

# Scatter Plot

```python
insurance_data  = pd.read_csv('./Data/insurance.csv')
insurance_data.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges)
plt.xlabel('bmi')
plt.ylabel('charges')
```

### To add regression line to the scatter plot

```python
plt.figure(figsize=(10,6))
sns.regplot(x=insurance_data.bmi, y=insurance_data.charges)
```

## Categorical Scatter Plot


```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = insurance_data , x="bmi", y="charges", hue="smoker", style="smoker")
```

### To add regression lines to a categorical scatter plot

```python
plt.figure(figsize=(10,6))
sns.lmplot(data = insurance_data , x="bmi", y="charges", hue="smoker")
```
