```python
import pandas as pd
wine_review = pd.read_csv('./winemag-data-130k-v2.csv')
wine_review.head()
```

## Groupwise analysis : groupby()

```python
wine_review.points.describe()
```

```python
wine_review.points.value_counts()
```

We can use groupby() to categorize data

```python
#To replicate value_counts() function
wine_review.groupby('points').points.count()
```

```python
#Cheapest wine in each point
wine_review.groupby('points').price.min()
```

```python
#selecting the name of the first wine rviewed from each winary
wine_review.groupby('winery').apply(lambda t: t.title.iloc(0))
```

```python
# groupby more than one feature
#Pick out the best wine by country and province
wine_review.groupby(['country','province']).apply(lambda ch: ch.points.max())
```

## To run bunch of different function on your grouped Dataframe simultanously: agg()

```python
wine_review.groupby(['country']).price.agg([len,min,max])
```

## Sorting
We can sort the (grouped)data with respect to index (sort_index()), or with respect to the values (sort_values())

```python
countries_reviewed = wine_review.groupby('country').description.agg([len])
```

```python
countries_reviewed.sort_values(by='len',ascending=False)
```

```python
countries_reviewed.sort_index()
```
