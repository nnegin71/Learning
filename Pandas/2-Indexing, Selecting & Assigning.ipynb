## Indexing in pandas

#### (1) Index-based selection, iloc
Selecting data based on its numerical position in the data. Columns are pointed to as index, not by their names.

Both iloc and loc are row first, column second commands.

```python
#To select the first row of data
wine_review.iloc[0]
```

```python
#row first, column second
wine_review.iloc[:,0]
```

```python
#To select first 3 rows of one column --->(0:3 means from zero, up to (but not includeing) 3)
wine_review.iloc[:3,0]
```

```python
#To select first rows 2nd and 3rd of one column
wine_review.iloc[1:3,0]
```

```python
#To select rows/columns with passing a list
wine_review.iloc[[1,2],0]
```

```python
#To select LAST 5 element of the column, use negative number
wine_review.iloc[-5:,0]
```

#### (2) Label-based selection, loc
In this method, the data index value matters not its position

```python
#To select specefic entry
wine_review.loc[0, 'country']
```

```python
#To select a few special columns of DataFrame
wine_review.loc[:,['taster_name','points']]
```

```python
#To select special rows of some special columns of DataFrame
wine_review.loc[:3,['taster_name','points']]
```

### Important note
iloc: Does the counting like python. 0:3 mean from 0 up to but not including 3

loc: Has inclusive counting method. 0:3 means from 0 to and including 3 . Because loc access the element by their labels, 'a':'z' means everything from 'a' to 'z'.

### Conditional selection

```python
# To find part of data that satisfy some conditions: batter than average Italy wines
wine_review.loc[(wine_review['country']=='Italy') & (wine_review['points']>=99)]
```

### isin ... Select data whose values is within a list

```python
# To select wines of only Italy and France
wine_review.loc[wine_review.country.isin(['Italy','France'])]
```

### isnull  ... Detect missing values
### notnull ... Select available data while filtering the missing ones

```python
wine_review.loc[wine_review.country.isnull()]
# wine_review.loc[wine_review.price.notnull()]
```
