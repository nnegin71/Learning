```python
import pandas as pd
import numpy as np

#plotting modules
import matplotlib.pyplot as plt
import seaborn as sns

#minmax scaling module
from mlxtend.preprocessing import minmax_scaling

# for Box-Cox Transformation
from scipy import stats

# set seed for reproducibility
np.random.seed(0)
```

## Scaling vs. Normalization
Scaling: Chaning the range of your data

        You want to scale data when you use the methods based on measures of how far apart data points are (K-nearest neighbor)
        
Normalization: Changing the shape of the distribution in your data

        In general, you'll normalize your data if you're going to be using a machine learning or statistics technique that assumes your data is normally distributed.

### Scaling


```python
#Generate 1000 random data points from an exponentil distribution
original_data = np.random.exponential(size=1000)
original_data.shape
```

```python
# Scale data between [0 1] - using mixmax_scaling 
#columns=0: scale each column independently 
#columns=1: scale each row independently
scaled_data = minmax_scaling(original_data, columns=0, min_val=0, max_val=1)
```

```python
#Plotting
fig,ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0],kde=True,legend=False)
ax[0].set_title("original data")
sns.histplot(scaled_data,ax=ax[1],kde=True,legend=False)
ax[1].set_title("scaled_data")
plt.show()
```

### Normalization

```python
#Normalizing data
normalized_data = stats.boxcox(original_data)

#plotting
fig,ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0],kde=True,legend=False)
ax[0].set_title("original data")
sns.histplot(normalized_data,ax=ax[1],kde=True,legend=False)
ax[1].set_title("normalized_data")
plt.show()
```
