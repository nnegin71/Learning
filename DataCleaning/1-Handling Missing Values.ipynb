```python
import numpy as np
import pandas as pd

nfl_v4=pd.read_csv('./NFL Play by Play 2009-2017 (v4).csv', index_col='Date')
```

```python
nfl_v4.head()
```

```python
#Getting the number of missed values
missing_values_count = nfl_v4.isnull().sum()
```

```python
print(missing_values_count)
```

```python
#Get total percentage of missed values
nfl_v4.shape
total_cell_count = np.product(nfl_v4.shape)
total_missing_count = missing_values_count.sum()
total_percent_missing = (total_missing_count/total_cell_count)*100
print(total_percent_missing)
```

### Figure out why some values are missed?
Is it because the missing value was not recorded OR it doesn't exist? You need to read data documentation to learn about data set

If the missing value doesn't exist, like the height of older child for a family without children. You need to keep that value as "NAN".

If the missing value is a missing measurement, you need to try guessing the value. It's called imputation.

#### 1. Drop missing values (dropna())

```python
#To remove the rows that have missed values
nfl_v4.dropna()
```

```python
#To remove the columns that have missed values
columns_with_na_dropped = nfl_v4.dropna(axis=1)
print("Number of dropped columns are: " , nfl_v4.shape[1]-columns_with_na_dropped.shape[1])
```

#### 2. Filling in missing values automatically: with a constant number or with whatever value comes directly after in the same column(fillna())

```python
nfl_v4_subset = nfl_v4.loc[:,'EPA':'WPA']
nfl_v4_subset.head()
```

```python
nfl_v4_subset.fillna(0).head()
```

Other `methods` that can be used with `fillna`:
* **bfill :**
  
  Backward fill: Uses the next valid value below the missing value to fill it.
  Fills `NaN` with the next known non-null value below it in the same column (if axis=0).

* **ffill:**
  
  Forward fill: Propagates the last valid value forward to fill gaps.
  Fills `NaN` with the last known non-null value above it in the same column (if axis=0).

```python
nfl_v4_subset.fillna(method='bfill',axis=0).fillna(0)
```
